```python
import pandas as pd
import numpy as np
s = pd.Series(range(5),index=list('EDCAB'))
```

```python
#Lets sort for index
s.sort_index()
```

```python
df1 = pd.DataFrame(np.arange(9).reshape(3,3),columns = ['Elmir','Orxan','Amin'],index=list('CAB'))
```

```python
df1.sort_index()
```

```python
df1.sort_index(axis=1) #sort for columns
```

```python
# If we want sorting in descending order, just use : ascending = False
df1.sort_index(axis=1,ascending=False)
```

#### Lets sort for values

```python
s1 = pd.Series([7,11,43],index= list('BCA'))
```

```python
s1.sort_values(ascending=False)
```

```python
#Multi and single sorting in DataFrame

df2= pd.DataFrame({'A':[1,2,34,23],'B':[87,43,75,-423]})
df2
```

```python
df2.sort_values(by='B',ascending=False)
```
